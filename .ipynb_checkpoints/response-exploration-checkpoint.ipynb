```python
def get_cycle(cmap, N=None, use_index="auto"):
    if isinstance(cmap, str):
        if use_index == "auto":
            if cmap in ['Pastel1', 'Pastel2', 'Paired', 'Accent',
                        'Dark2', 'Set1', 'Set2', 'Set3',
                        'tab10', 'tab20', 'tab20b', 'tab20c']:
                use_index=True
            else:
                use_index=False
        cmap = plt.get_cmap(cmap)
    if not N:
        N = cmap.N
    if use_index=="auto":
        if cmap.N > 100:
            use_index=False
        elif isinstance(cmap, LinearSegmentedColormap):
            use_index=False
        elif isinstance(cmap, ListedColormap):
            use_index=True
    if use_index:
        ind = np.arange(int(N)) % cmap.N
        return cycler("color",cmap(ind))
    else:
        colors = cmap(np.linspace(0,1,N))
        return cycler("color",colors)
```

```python
## dapting this from hasasia. With efac=1.0, and equad/ecorr set to 0, this correlation matrix should just be diagonal in efac**2
## because this dataset has no detailed treatment of individual observatories, we ignore all the backend-specificity of the original code
def make_corr(psr):
    N = psr.toaerrs.size
    corr = np.zeros((N,N))

    sigma_sqr = np.zeros(N)

    efac = 1.0
    sigma_sqr = efac**2 * psr.toaerrs**2
    
    corr = np.diag(sigma_sqr)
    return corr
```

Define all the echo functions and the binary class:

```python
## definitions for f(t) and t(f)

def dt_by_f(f_gw,mc,units_out=u.yr):
    '''
    Returns delta_t = t_c - t (time to coalescence t_c)
    
    Assumes f_gw in Hz, mc in Msun.
    
    units_out (astropy.unit) must be a time unit; result will be returned in these units (default yrs)
    '''
    ## unit checking
    if not hasattr(f_gw,'unit'):
        f_gw = f_gw*u.Hz
    elif f_gw.unit is not u.Hz:
        f_gw = f_gw.to(u.Hz)
    
    if not hasattr(mc,'unit'):
        mc = mc*u.M_sun
        mc = mc.to(u.kg)
    elif mc.unit is not u.kg:
        mc = mc.to(u.kg)
    
    return ((5/(8*np.pi)**(8/3)) * (c.G*mc/c.c**3)**(-5/3) * f_gw**(-8/3)).decompose().to(units_out)

def f_by_dt(dt,mc,units_in=u.yr):
    '''
    Inverse of above. dt = t_c - t (i.e., time to coalescence)
    '''
    if not hasattr(dt,'unit'):
        dt = dt*units_in.to(u.s)
    elif dt.unit is not u.s:
        dt = dt.to(u.s)
    
    if not hasattr(mc,'unit'):
        mc = mc*u.M_sun
        mc = mc.to(u.kg)
    elif mc.unit is not u.kg:
        mc = mc.to(u.kg)
    
    return ((5**(3/8) / (8*np.pi)) * (c.G*mc/c.c**3)**(-5/8) * dt**(-3/8)).decompose().to(u.Hz)

```

Largely following Hazboun+2019 (https://arxiv.org/pdf/1907.04341) eqn. 82 and surrounding derivations:

```python
## strain amplitude for a monochromatic source h_echo
def calc_h_echo(mc,f_gw,d_L):

    ## deal with units
    if not hasattr(mc,'unit'):
        mc = mc*u.M_sun
        mc = mc.to(u.kg)
    elif mc.unit is not u.kg:
        mc = mc.to(u.kg)

    if not hasattr(f_gw,'unit'):
        f_gw = f_gw*u.Hz
    elif f_gw.unit is not u.Hz:
        f_gw = f_gw.to(u.Hz)

    if not hasattr(d_L,'unit'):
        d_L = d_L*u.Mpc
        d_L = d_L.to(u.m)
    elif f_gw.unit is not u.m:
        d_L = d_L.to(u.m)
    
    h0 = (4*c.c/d_L) * (c.G*mc/c.c**3)**(5/3) * (np.pi*f_gw)**(2/3)

    return h0.decompose()
```

```python
## let's make a class for the binaries
class MBHB:

    def __init__(self,mc,d_L,sky_loc=None,circular=True):
        '''
        Class to house binary attributes and related methods.

        Arguments
        --------------------
        mc (float or astropy.Quantity):
            Chirp mass. Assumed to be in solar masses if no unit is provided
        d_L (float or astropy.Quantity):
            Luminosity distance. Assumed to be in Mpc if no unit is provided
        sky_loc (astropy.coordinates.SkyCoord object):
            Sky location of the merger as seen from Earth, as an astropy SkyCoord object.
        circular (bool):
            Whether to treat the binary as circular (Default True; eccentric binaries not yet implemented)
        
        '''
        ## we may eventually wish to have eccentric systems
        if not circular:
            raise NotImplementedError

        ## deal with units
        if not hasattr(mc,'unit'):
            self.mc = mc*u.M_sun
        elif mc.unit is not u.M_sun:
            self.mc = mc.to(u.M_sun)
        else:
            self.mc = mc
    
        if not hasattr(d_L,'unit'):
            self.d_L = d_L*u.Mpc
        elif d_L.unit is not u.m:
            self.d_L = d_L.to(u.Mpc)
        else:
            self.d_L = d_L

        self.sky_loc = sky_loc

        ## call the merger frequency the frequency at t_c - t = 1 s
        ## NOTE: we diverge from the Hazboun+19 convention here; their f0 is the monochromatic PTA-band frequency,
        ##       whereas ours is the frequency at merger in the LISA band
        self.f0 = f_by_dt(1*u.s,self.mc)

    ## wrappers for f(t), t(f), h_echo(f), h_echo(t)
    def f_of_t(self,t):
        return f_by_dt(t,self.mc)
    def t_of_f(self,f):
        return(f,self.mc)
    def h_echo_of_f(self,f):
        return calc_h0(self.mc,f,self.d_L)
    def h_echo_of_t(self,t):
        foft = self.f_of_t(t)
        return calc_h_echo(self.mc,foft,self.d_L)
```

```python
## SNRs
## various averaging terms, response, etc. are absorbed into S_eff by hasasia
def calc_snr(h0,T_obs,S_eff):

    return h0*np.sqrt(T_obs/S_eff)

## array SNR
## I eventually want to build an PhasedArray class to handle a bunch of the setup and call this method
## as only a function of the binary object but this is fine for now
def echo_array_snr(binary,psr_specs,psr_dists=None,nside=32):

    ## hasasia skysens takes (theta_gw,phi_gw), following the Healpy pix2ang convention
    ## but we have a SkyCoord that doesn't use the silly Healpy convention
    ## so convert from longitude and latitude into colatitude and longitude
    theta_gw = np.asarray(np.abs(binary.sky_loc.barycentricmeanecliptic.lat.rad - np.pi/2)).reshape(1)
    phi_gw = np.asarray(binary.sky_loc.barycentricmeanecliptic.lon.rad).reshape(1)

    if psr_dists is None:
        psr_dists = [psr_spec.pdist for psr_spec in psr_specs]

    ## deal with units
    if not hasattr(psr_dists[0],'unit'):
        print("Warning, no unit provided for pulsar distances. Assuming kpc.")
        psr_dists = [psr_dist*u.kpc for psr_dist in psr_dists]
    
    snrs_squared = []
    psr_gw_freqs = []
    psr_gw_amps = []
    for psr_spec, psr_dist in zip(psr_specs, psr_dists):
        ## make the directional responses
        sky_spec = hsky.SkySensitivity([psr_spec],theta_gw, phi_gw, pulsar_term='only')
        ## get Tspan
        T = sky_spec.Tspan

        ## Use pulsar attributes to determine GW frequency in pulsar
        time_delay = (1+np.einsum('ij,il->jl',sky_spec.pos,sky_spec.K)) * psr_dist/c.c
        import pdb; pdb.set_trace()
        pulsar_freq = binary.f_of_t(time_delay)
        psr_gw_freqs.append(pulsar_freq)
        ## get frequency array index of GW frequency in this pulsar
        pf_idx = np.argmin(np.abs(psr_spec.freqs-pulsar_freq.value))

        ## deal with high frequencies
        last_bin_width = psr_spec.freqs[-1] - psr_spec.freqs[-2]
        if (pf_idx == len(psr_spec.freqs)-1) and (pulsar_freq.value > psr_spec.freqs[-1]+last_bin_width):
            S_eff = np.inf
        else:
            ## effective noise curve
            S_eff = sky_spec.S_eff[pf_idx]

        ## get monochromatic strain at t = -time_delay
        psr_gw_amp = binary.h_echo_of_t(time_delay)
        psr_gw_amps.append(psr_gw_amp)

        ## compute snr squared
        snr_squared = psr_gw_amp**2  * (T/S_eff)
        
        snrs_squared.append(snr_squared)

    return snrs_squared, psr_gw_freqs, psr_gw_amps
        
        
```

Do all the data preliminaries. For complete explanations, see sensitivity-estimates.ipynb.

```python
rnoise = {}

for nf in noise_files:
    with open(nf,'r') as fin:
        rnoise.update(json.load(fin))

## load pulsar instances
ePsrs = []
for par,tim in zip(pars,tims):
    ePsr = ePulsar(par, tim,  ephem='DE436')
    ePsrs.append(ePsr)
    print('\rPSR {0} complete'.format(ePsr.name),end='',flush=True)

## get timespan
Tspan = hsen.get_Tspan(ePsrs)

print("Timespan:{:0.1f} yrs".format(Tspan/(3600*24*365))) ## in yrs

## frequencies
fyr = 1/(365.25*24*3600)
freqs = np.logspace(np.log10(1/(5*Tspan)),np.log10(1e-6),800)

## hasasia Spectrum instances
psrs = []
thin = 10 
os.environ['JAX_PLATFORMS'] = 'cpu'
for ePsr in ePsrs:
    corr = make_corr(ePsr)[::thin,::thin]
    plaw = hsen.red_noise_powerlaw(A=9e-16, gamma=13/3., freqs=freqs)
    if ePsr.name in rnoise.keys():
        logAmp, gam = rnoise[ePsr.name]['log10_A'], rnoise[ePsr.name]['gamma']
        plaw += hsen.red_noise_powerlaw(A=10**logAmp, gamma=gam, freqs=freqs)
        
    corr += hsen.corr_from_psd(freqs=freqs, psd=plaw,
                               toas=ePsr.toas[::thin])
    psr = hsen.Pulsar(toas=ePsr.toas[::thin],
                      toaerrs=ePsr.toaerrs[::thin],
                      phi=ePsr.phi,theta=ePsr.theta, 
                      N=corr, designmatrix=ePsr.Mmat[::thin,:])
    psr.name = ePsr.name
    psrs.append(psr)
    # del ePsr
    print('\rPSR {0} complete'.format(psr.name),end='',flush=True)
```

```python
## run on a grid
result_dict = {}
masses = [1e7,5e7,1e8,5e8,1e9,5e9,1e10] ## in Msun
merger_dists = [1,2.5,5,10] ## in Gpc

for m in masses:
    for dl in merger_dists:
        name = "{:0.0e}Msun_{}Gpc".format(m,dl)
        result_dict[name] = {"Mc":m*u.Msun,
                             "mc_name":"{:0.0e}".format(m)+" $\\mathrm{M}_{\\odot}$",
                             "dl_name":"{} Gpc".format(dl),
                             "D_L":dl*1000*u.Mpc}
        
```

```python
nside=16
run_HP = ahp.HEALPix(nside,frame='barycentricmeanecliptic')
```

```python
## run the grid
for i, run in enumerate(result_dict.keys()):
    print("Beginning run {}/{}: {}".format(i+1,len(result_dict),run))
    ## instantiate the binary
    binary_i = MBHB(mc=result_dict[run]["Mc"],d_L=result_dict[run]["D_L"])
    ## keep track of quantities of interest
    echo_fs_i = []
    echo_amps_i = []
    rho2_map_i = []
    for idx in tqdm(range(hp.nside2npix(nside))):
        sky_loc_i = run_HP.healpix_to_skycoord(idx)
        binary_i.sky_loc = sky_loc_i
        rho2_i, fs_i, amps_i = echo_array_snr(binary_i,specs,nside=nside)
        rho2_map_i.append(rho2_i)
        echo_fs_i.append(fs_i)
        echo_amps_i.append(amps_i)

    result_dict[run]["echo_fs"] = echo_fs_i
    result_dict[run]["echo_amps"] = echo_amps_i
    result_dict[run]["echo_rho2s"] = rho2_map_i
    
    ## coherently add the SNR^2 across pulsars, excluding nans
    summed_rho2_map_i = [np.sum(np.array(rho2_i)[np.invert(np.isnan(rho2_i))]) for rho2_i in rho2_map_i]

    result_dict[run]["snr_squared_map"] = summed_rho2_map_i
    
    plot_title = "Echo Array SNR by GW Origin: {} at {}".format(result_dict[run]["mc_name"],
                                                                result_dict[run]["dl_name"])
    result_dict[run]['title'] = plot_title
    hp_old.mollview(np.sqrt(summed_rho2_map_i),unit="SNR",title=plot_title)
    hp_old.projscatter(theta=[sp.theta for sp in specs],phi=[sp.phi for sp in specs],
                      s=3,marker='*',c='cyan')
    result_dict[run]["snr_map_fig"] = plt.gcf()
    plt.savefig("./figures/DR3like_{}_SNR_map.png".format(run))
    plt.close()
    
```
